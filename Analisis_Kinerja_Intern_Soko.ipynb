# **Dashboard Kinerja Intern Soko Financial**


```python
# Filter the DataFrame based on 'Keterangan' column
filtered_df6 = df[(df['Keterangan'] == 'Tidak Aktif') | (df['Keterangan'] == 'Resign') | (df['Keterangan'] == 'Aktif')]

# Group by 'Divisi' and 'Keterangan' and get counts
divisi_keterangan_counts = filtered_df6.groupby(['Divisi', 'Keterangan'])['Keterangan'].count().reset_index(name='Count')

# Define colors for each category
colors = {
    'Aktif': 'mediumslateblue',
    'Tidak Aktif': 'red',
    'Resign': 'grey',
}

# Create the grouped bar chart
plt.figure(figsize=(10, 6))  # Adjust figure size if needed
sns.barplot(x='Divisi', y='Count', hue='Keterangan', data=divisi_keterangan_counts, palette=colors)
plt.title('Distribusi Keterangan untuk Setiap Divisi')
plt.xlabel('Divisi')
plt.ylabel('Jumlah')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show() # To see the output, run the code
```

```python
# Filter the DataFrame based on 'Keterangan' column
filtered_df7 = df[(df['Keterangan'] == 'Tidak Aktif') | (df['Keterangan'] == 'Resign') | (df['Keterangan'] == 'Aktif')]

# Group by 'Divisi' and 'Keterangan' and get counts
divisi_keterangan_counts = filtered_df7.groupby(['Divisi', 'Keterangan'])['Keterangan'].count().reset_index(name='Count')

# Define colors for each category
colors = {
    'Aktif': 'mediumslateblue',
    'Tidak Aktif': 'red',
    'Resign': 'grey',
}

# Calculate the number of rows and columns for subplots
num_divisi = divisi_keterangan_counts['Divisi'].nunique()
num_cols = 3  # Number of columns you want
num_rows = (num_divisi + num_cols - 1) // num_cols  # Calculate rows dynamically

# Create a figure and axes for subplots with dynamic sizing
fig, axes = plt.subplots(num_rows, num_cols, figsize=(12, 4 * num_rows))  # Adjust figure size as needed
axes = axes.flatten()  # Flatten the axes array for easier iteration

# Iterate through each divisi and create a pie chart
for i, (divisi, group) in enumerate(divisi_keterangan_counts.groupby('Divisi')):
    ax = axes[i]  # Get the current subplot axes

    # Create the pie chart
    wedges, texts, autotexts = ax.pie(
        group['Count'],
        labels=group['Keterangan'],
        colors=[colors.get(label, 'gray') for label in group['Keterangan']],
        autopct='%1.1f%%',
        startangle=90,
        wedgeprops=dict(width=0.5),
    )

    # Add the count as text to each slice
    for j, autotext in enumerate(autotexts):
        autotext.set_text(f'{autotext.get_text()} ({group["Count"].iloc[j]})')

    ax.set_title(f'Divisi {divisi}')
    ax.axis('equal')

# Hide any unused subplots
for i in range(num_divisi, num_rows * num_cols):
    axes[i].axis('off')

# Adjust layout and display the plot
plt.tight_layout()
plt.show() # To see the output, run the code
```
